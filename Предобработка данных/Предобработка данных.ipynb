### Шаг 2.1 Заполнение пропусков

```python
display(data[data['total_income'].isna()])
```

<div class="alert alert-info"> <b>Комментарий студента:</b> 
    Таким образом, мы видим, что в стобцах days_employed и total_income есть пропущенные значения NaN. Такие пропуски принадлежат к типу данных float, то есть с ними можно проводить арифметические операции. Однако такие данные, как количество дней проработанных на работе и суммарный доход могут напрямую влиять на анализ, то лучше пропуски не удалять, а заменить их медианными значениями. </div>

```python
nan_part_in_total_income = data['total_income'].isna().sum()/len(data['total_income'])
```

```python
print('Доля пропусков в столбце total_income', nan_part_in_total_income)
```

```python
nan_part_in_days_employed = data['days_employed'].isna().sum()/len(data['days_employed'])
```

```python
print('Доля пропусков в столбце days_employed', nan_part_in_days_employed)
```

<div class="alert alert-info"> <b>Комментарий студента:</b> 
    Доля пропусков для двух столбцов одинаковая. Пенсионеры, госслужащие получают точно официально заработную плату|пенсию. Поэтому данная ошибка больше похожа на механическую ошибку при получении данных, нежели на отсутствие официальной работы или отсутствие какого-либо дохода. Т.к. исходя из этого можно подумать, что пропуски не являются отсутсвием дохода у людей, а также работы, нельзя просто заменить значения на 0. При этом удалить строки с пропусками тоже нельзя, так как это окажет влияние на исследование, так как пропуски соствляют около 10% от всего объема. В таком случае лучше заменить пропуски медианным значением, что не окажет сильного влияния на итоги исследования.</div>
